## Top 15 States with highest accident rates an find which State has maximum number of accidents

## Timestamp

```python
type(df.Start_Time)# chnge the tpe firt it was pandas.core.series.Series now it is 
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time)
df.Start_Time
```

```python
df['Month'] = df['Start_Time'].dt.month
df['Year'] = df['Start_Time'].dt.year
df['Hour'] = df['Start_Time'].dt.hour
df['Weekday'] = df['Start_Time'].dt.weekday
dt_16=df[df.Start_Time.dt.year==2016]
dt_17=df[df.Start_Time.dt.year==2017]
dt_18=df[df.Start_Time.dt.year==2018]
dt_19=df[df.Start_Time.dt.year==2019]
```

## Yearly

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Year", data=df, orient = 'v', palette = "crest").set_title("Number of Accidents Years")
```

### There is rapid increase in number of accidents in the year 2021 compared to rest of years.

# 

## Monthly

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Month", data=dt_16, orient = 'v', palette = "crest").set_title("Number of Accidents in Month of Year 2016")
```

### It can be observed that the number of accidents gradually increased from January to December in the year 2016. The analysis shows that December had the highest number of accidents among all the months in 2016

# 

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Month", data=dt_17, orient = 'v', palette = "crest").set_title("Number of Accidents in Month of Year 2017")
```

### It can be observed that the trend of increasing accidents from the end of 2016 continued into the year 2017, and then decreased after June. Specifically, the analysis shows that the month of March in 2017 had the highest number of accidents compared to all other months in that year

# 

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Month", data=dt_18, orient = 'v', palette = "crest").set_title("Number of Accidents in Month of Year 2018")
```

### Number of accidents till June kept decreasing but after June they gradually inceased. December has highest number of accidents

# 

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Month", data=dt_19, orient = 'v', palette = "crest").set_title("Number of Accidents in Month of Year 2019")
```

### Number of accidents start increasing from September to December. December has highest number of accidents

## Hourly 

##  During which hours of the day maximum accidents occured and what are the hours range that are prone to accidents?

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Hour", data=df, orient = 'v', palette = "icefire_r")
```

###  It can be observed that the number of accidents tends to peak during the morning hours between 5am to 9am and then gradually decreases until it starts to peak again around 1pm to 6pm. Furthermore, it can be inferred that rush hour traffic during those hours in the morning to commute to work leads to more traffic and increased chances of accidents. This inference is supported by common knowledge of traffic patterns during morning and evening rush hour traffic.
